```python
import math
import random

system_1 = readCoNLL("/Users/slouvan/sandbox/emnlp2017-bilstm-cnn-crf/results/SingleTask_ATIS_200/predictions/ATIS_34_test.conll", {0:'tokens', 1 : 'gold', 2 : 'prediction'})
system_2  = readCoNLL("/Users/slouvan/sandbox/emnlp2017-bilstm-cnn-crf/results/MultiTask_ATIS_200_OntoNotes_NW/predictions/ATIS_test.conll", {0 :'tokens', 1 :'gold', 2:'prediction'})


def check_significance(file_1, file_2, trials):
    system_1 = readCoNLL(file_1, {0:'tokens', 1 : 'gold', 2 : 'prediction'})
    system_2  = readCoNLL(file_2, {0 :'tokens', 1 :'gold', 2:'prediction'})
    gold = [t['gold'] for t in system_1]
    system_1_ori = [t['prediction'] for t in system_1]
    system_2_ori = [t['prediction'] for t in system_2]

    _, _, f1_system_1 = do_computeF1ScoreCONLL(gold, system_1_ori)
    print("Real F1 System 1 : {}".format(f1_system_1))
    _, _, f1_system_2 = do_computeF1ScoreCONLL(gold, system_2_ori)
    print("Real F1 System 2 : {}".format(f1_system_2))
    
    absolute_difference = math.fabs(f1_system_1 - f1_system_2)
    print("Absolute difference :  {}".format(absolute_difference))
    at_least_count = 0
    for i in range(0, trials) :
        pseudo_1_all = []
        pseudo_2_all = []
        for j in range(0, len(gold)):
            pseudo_1_current = []
            pseudo_2_current = []
            for k in range(0, len(gold[j])): 
                if random.randint(0, 1) == 0:
                    pseudo_1_current.append(system_1_ori[j][k])
                    pseudo_2_current.append(system_2_ori[j][k])
                else :
                    pseudo_1_current.append(system_2_ori[j][k])
                    pseudo_2_current.append(system_1_ori[j][k])
            pseudo_1_all.append(pseudo_1_current)
            pseudo_2_all.append(pseudo_2_current)

        _, _, f1_pseudo_system_1 = do_computeF1ScoreCONLL(gold, pseudo_1_all)
        _, _, f1_pseudo_system_2 = do_computeF1ScoreCONLL(gold, pseudo_2_all)

        pseudo_difference = math.fabs(f1_pseudo_system_1 - f1_pseudo_system_2)
        if pseudo_difference >= absolute_difference :
            at_least_count += 1
            #print(at_least_count / i * 1.0)

    print( (at_least_count + 1.0) / (trials + 1.0) )
```
