```python
def filter_tag(input_file_name, output_file_name, relevant_tag):
    # read file file_name into sentences
    original_sentences = readCoNLL(input_file_name,{0:'tokens', 1:'GOLD', 2:'PREDICT'})
    
    # for each sentence filter out irrelevant tag
    for idx, sentence in enumerate(original_sentences):
        gold_labels = sentence['GOLD']
        predicted_labels = sentence['PREDICT']
        for gold_idx, label in enumerate(gold_labels):
            label_name = get_label_name(label)
            if label_name not in relevant_tag or label_name not in relevant_tag :
                gold_labels[gold_idx] = "O"
        for predicted_idx, label in enumerate(predicted_labels):
            label_name = get_label_name(label)
            if label_name not in relevant_tag or label_name not in relevant_tag :
                predicted_labels[predicted_idx] = "O"
        original_sentences[idx]['GOLD'] = gold_labels
        original_sentences[idx]['PREDICT'] = predicted_labels

    # output to output_file_name
    with open(output_file_name, "w") as f:
        for idx, sentence in enumerate(original_sentences):
            tokens, gold_labels, predicted_labels = sentence['tokens'], sentence['GOLD'], sentence['PREDICT']
            for idx, _ in enumerate(tokens):
                print("{} {} {}".format(tokens[idx], gold_labels[idx], predicted_labels[idx]))
                f.write("{} {} {}\n".format(tokens[idx], gold_labels[idx], predicted_labels[idx]))
            f.write("\n")
    
```

```python
def load_tag_from_mapping(input_file_name, conll_type):
    relevant_tag = []
    with open(input_file_name, "r") as f:
        lines = f.readlines()
    
    for line in lines:
        slot_tag, conll_tag = line.split(",")
        #print("{} {}".format(slot_tag, conll_tag))
        if conll_tag.strip() == conll_type:
            relevant_tag.append(slot_tag)
    
    return relevant_tag
```
